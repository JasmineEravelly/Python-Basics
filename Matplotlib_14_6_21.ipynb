```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Sample Graph in 2D')
```

```python
plt.plot(x,y,'r--*',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

- subplot

```python
y1=x**2
y2=y1**2
```

```python
plt.subplot(2,2,1)
plt.plot(x,y1,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y2,'g--')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave from")
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
plt.title("sine wave from")
plt.plot(x,y)
```

- Box plot using Matplotlib

```python
data= [np.random.normal(0,std, 100) for std in range(1,4)]
```

```python
plt.boxplot(data,vert=True,patch_artist=True);
```

- Pie Chart

```python
labels='Python','Java','Ruby','C++'
sizes=[215,180,150,200]
colors=['gold','yellowgreen','lightcoral','skyblue']
explode=(0.4,0,0,0.2)

plt.pie(sizes,colors=colors,labels=labels,explode=explode,shadow=True,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```
